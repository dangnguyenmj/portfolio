```python
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, r2_score, mean_absolute_percentage_error
from sklearn.model_selection import TimeSeriesSplit
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
import os
import joblib 
```

```python
# Danh sách mã cổ phiếu
tickers = ["XOM", "CVX", "COP", "NEE", "ENB", "SLB", "KMI", "EOG","SHEL", "TTE"]

end = datetime.now()
start = "2015-01-01"
```

```python
# Thư mục lưu mô hình
MODEL_DIR = "../models"
if not os.path.exists(MODEL_DIR):
    os.makedirs(MODEL_DIR)
```

```python

for stock in tickers:
    print(f"Đang xử lý mã cổ phiếu: {stock}")
    data = yf.download(stock, start=start, end=end)
    if data.empty:
        print(f"Không có dữ liệu cho {stock}, bỏ qua.")
        continue

    # Vẽ đồ thị giá đóng cửa điều chỉnh
    plt.figure(figsize=(15, 5))
    plt.plot(data.index, data['Adj Close'], label='Adj Close')
    plt.xlabel("Years")
    plt.ylabel("Adj Close")
    plt.title(f"Closing Price of {stock} Data")
    plt.legend()
    plt.show()

    # Tính và vẽ MA (Moving Average)
    data['MA_for_250_days'] = data['Adj Close'].rolling(250).mean()
    data['MA_for_100_days'] = data['Adj Close'].rolling(100).mean()

    plt.figure(figsize=(15, 5))
    plt.plot(data.index, data['Adj Close'], label='Adj Close')
    plt.plot(data.index, data['MA_for_100_days'], label='MA_for_100_days', color='orange')
    plt.plot(data.index, data['MA_for_250_days'], label='MA_for_250_days', color='green')
    plt.xlabel("Years")
    plt.ylabel("Price")
    plt.title(f"{stock} Adj Close with 100 & 250 days MA")
    plt.legend()
    plt.show()

    # Tiền xử lý dữ liệu
    Adj_close_price = data[['Adj Close']].dropna()
    if len(Adj_close_price) < 500:
        print(f"Dữ liệu cho {stock} quá ít ({len(Adj_close_price)} data points), bỏ qua.")
        continue

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(Adj_close_price)
    
    # Chuẩn bị dữ liệu cho LSTM (10 timesteps)
    n_steps_lstm = 10 
    x_data, y_data = [], []
    for i in range(n_steps_lstm, len(scaled_data)):
        x_data.append(scaled_data[i - n_steps_lstm:i])
        y_data.append(scaled_data[i])
    x_data, y_data = np.array(x_data), np.array(y_data)

    # Chia dữ liệu train/test
    splitting_len = int(len(x_data) * 0.7)
    x_train, y_train = x_data[:splitting_len], y_data[:splitting_len]
    x_test, y_test = x_data[splitting_len:], y_data[splitting_len:]

    # Cross-validation trên train set
    tscv = TimeSeriesSplit(n_splits=5) 
    fold_rmse = []
    fold_mae = []
    fold_mape = []
    fold_r2 = []

    # Tiến hành cross-validation
    for train_index, val_index in tscv.split(x_train):
        X_train_cv, X_val_cv = x_train[train_index], x_train[val_index]
        y_train_cv, y_val_cv = y_train[train_index], y_train[val_index]

        model = Sequential()
        model.add(LSTM(120, return_sequences=True, input_shape=(X_train_cv.shape[1], 1)))
        model.add(Dropout(0.5))
        model.add(LSTM(64, return_sequences=False))
        model.add(Dropout(0.5))
        model.add(Dense(25))
        model.add(Dense(1))

        model.compile(optimizer=Adam(learning_rate=0.0005), loss='mean_squared_error')
        early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        model.fit(X_train_cv, y_train_cv, batch_size=32, epochs=20, 
                  validation_data=(X_val_cv, y_val_cv), callbacks=[early_stop], verbose=0)

        val_predictions = model.predict(X_val_cv)
        inv_val_predictions = scaler.inverse_transform(val_predictions)
        inv_y_val_cv = scaler.inverse_transform(y_val_cv)

        rmse_val = np.sqrt(np.mean((inv_val_predictions - inv_y_val_cv) ** 2))
        mae_val = mean_absolute_error(inv_y_val_cv, inv_val_predictions)
        mape_val = mean_absolute_percentage_error(inv_y_val_cv, inv_val_predictions)
        r2_val = r2_score(inv_y_val_cv, inv_val_predictions)

        fold_rmse.append(rmse_val)
        fold_mae.append(mae_val)
        fold_mape.append(mape_val)
        fold_r2.append(r2_val)

    # In ra kết quả CV
    print(f"{stock} Mean RMSE from CV: {np.mean(fold_rmse):.4f}")
    print(f"{stock} Mean MAE from CV: {np.mean(fold_mae):.4f}")
    print(f"{stock} Mean MAPE from CV: {np.mean(fold_mape):.4f}")
    print(f"{stock} Mean R² from CV: {np.mean(fold_r2):.4f}")

    # Huấn luyện mô hình cuối trên toàn bộ x_train, y_train
    model_final = Sequential()
    model_final.add(LSTM(120, return_sequences=True, input_shape=(x_train.shape[1], 1)))
    model_final.add(Dropout(0.5))
    model_final.add(LSTM(64, return_sequences=False))
    model_final.add(Dropout(0.5))
    model_final.add(Dense(25))
    model_final.add(Dense(1))

    model_final.compile(optimizer=Adam(learning_rate=0.0005), loss='mean_squared_error')
    early_stop_final = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

    # Dùng một phần x_train để validation (10%)
    val_ratio = 0.1
    val_size = int(len(x_train)*val_ratio)
    model_final.fit(x_train[:-val_size], y_train[:-val_size], 
                    batch_size=32, epochs=20, 
                    validation_data=(x_train[-val_size:], y_train[-val_size:]),
                    callbacks=[early_stop_final], verbose=0)

    # Dự đoán trên test set
    test_predictions = model_final.predict(x_test)
    inv_test_predictions = scaler.inverse_transform(test_predictions)
    inv_y_test = scaler.inverse_transform(y_test)

    rmse_test = np.sqrt(np.mean((inv_test_predictions - inv_y_test) ** 2))
    mae_test = mean_absolute_error(inv_y_test, inv_test_predictions)
    mape_test = mean_absolute_percentage_error(inv_y_test, inv_test_predictions)
    r2_test = r2_score(inv_y_test, inv_test_predictions)

    print(f"{stock} Test RMSE: {rmse_test:.4f}")
    print(f"{stock} Test MAE: {mae_test:.4f}")
    print(f"{stock} Test MAPE: {mape_test:.4f}")
    print(f"{stock} Test R²: {r2_test:.4f}")

    # Vẽ biểu đồ so sánh dự đoán và thực tế trên tập test
    ploting_data = pd.DataFrame({
        'Original Test Data': inv_y_test.reshape(-1),
        'Predictions': inv_test_predictions.reshape(-1)
    }, index=data.index[splitting_len + 10:])

    plt.figure(figsize=(15, 6))
    plt.plot(ploting_data.index, ploting_data['Original Test Data'], label='Actual')
    plt.plot(ploting_data.index, ploting_data['Predictions'], label='Predicted')
    plt.title(f"Comparison of Actual and Predicted Prices (Test Set) for {stock}")
    plt.xlabel("Time")
    plt.ylabel("Price")
    plt.legend()
    plt.show()

    # Dự đoán giá cho ngày tiếp theo với n_steps=10
    last_10_days = data['Adj Close'][-10:].values.reshape(-1, 1)
    scaled_last_10_days = scaler.transform(last_10_days)
    x_input = scaled_last_10_days.reshape(1, n_steps_lstm, 1)  
    predicted_price = model_final.predict(x_input)
    predicted_price = scaler.inverse_transform(predicted_price)
    print(f"{stock} Predicted price for the next day: ${predicted_price[0][0]:.2f}\n")


    # Lưu mô hình
    model_path = os.path.join(MODEL_DIR, f"LSTM_model_{stock}.keras")
    model_final.save(model_path)
```
